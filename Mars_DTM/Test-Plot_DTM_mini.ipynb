```python
def make_dvec(theta=np.pi/5):
    phi = np.random.uniform(-np.pi,np.pi)
    dvec = np.asarray([np.sin(theta)*np.cos(phi), 
                                     np.sin(theta)*np.sin(phi), 
                                     np.cos(theta)])
    return dvec
```

```python
def ortho_proj(V,p,d):
    #d = np.expand_dims(d,axis=0).T
    Q=V-p
    qdot = Q.dot(d)
    proj = (qdot * np.expand_dims(d, axis=1)).T
    oproj = Q-proj
    return oproj

def closest_vertex(V,p,d):
    op = ortho_proj(V,p,d)
    dist = np.linalg.norm(ortho_proj(V,p,d),axis=1)
    idx = np.argmin(dist)
    return V[idx]
```
